```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer  
df = pd.read_csv('Intermediate result1.csv')
print(df.head())  
```

```python
df = df.dropna(axis=0, how='all')  
df = df.dropna(axis=1, how='all') 
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
imputer_mean = SimpleImputer(strategy='mean') 
df[numeric_columns] = imputer_mean.fit_transform(df[numeric_columns])
categorical_columns = df.select_dtypes(include=['object']).columns
imputer_mode = SimpleImputer(strategy='most_frequent')
df[categorical_columns] = imputer_mode.fit_transform(df[categorical_columns])
print(df.isnull().sum())
df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['HSC GPA'] = label_encoder.fit_transform(df['HSC GPA'])
df['Proposed HSC  GPA-1'] = label_encoder.fit_transform(df['Proposed HSC  GPA-1'])
df['Proposed HSC GPA-2'] = label_encoder.fit_transform(df['Proposed HSC GPA-2'])
categorical_columns = df.select_dtypes(include=['object']).columns  
for column in categorical_columns:
    if column not in ['HSC GPA', 'Proposed HSC  GPA-1', 'Proposed HSC GPA-2']:
        df[column] = label_encoder.fit_transform(df[column])
X = df.drop(columns=['HSC GPA', 'Proposed HSC  GPA-1', 'Proposed HSC GPA-2'])
y1 = df['HSC GPA']  # Target 1
y2 = df['Proposed HSC  GPA-1']  # Target 2
y3 = df['Proposed HSC GPA-2']  # Target 3

# Step 5: Split the data into training and testing sets
X_train, X_test, y_train1, y_test1 = train_test_split(X, y1, test_size=0.2, random_state=42)
X_train, X_test, y_train2, y_test2 = train_test_split(X, y2, test_size=0.2, random_state=42)
X_train, X_test, y_train3, y_test3 = train_test_split(X, y3, test_size=0.2, random_state=42)

# Step 6: Train a Naive Bayes classifier for each target variable
nb1 = GaussianNB()
nb1.fit(X_train, y_train1)
y_pred1 = nb1.predict(X_test)

nb2 = GaussianNB()
nb2.fit(X_train, y_train2)
y_pred2 = nb2.predict(X_test)

nb3 = GaussianNB()
nb3.fit(X_train, y_train3)
y_pred3 = nb3.predict(X_test)

# Step 7: Evaluate the model's performance
print("Performance for HSC GPA:")
print(f"Accuracy: {accuracy_score(y_test1, y_pred1) * 100:.2f}%")
print("Classification Report:")
print(classification_report(y_test1, y_pred1))
print("Confusion Matrix:")
print(confusion_matrix(y_test1, y_pred1))

print("\nPerformance for Proposed HSC GPA:")
print(f"Accuracy: {accuracy_score(y_test2, y_pred2) * 100:.2f}%")
print("Classification Report:")
print(classification_report(y_test2, y_pred2))
print("Confusion Matrix:")
print(confusion_matrix(y_test2, y_pred2))

print("\nPerformance for Proposed HSC GPA2:")
print(f"Accuracy: {accuracy_score(y_test3, y_pred3) * 100:.2f}%")
print("Classification Report:")
print(classification_report(y_test3, y_pred3))
print("Confusion Matrix:")
print(confusion_matrix(y_test3, y_pred3))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, cohen_kappa_score, precision_score, recall_score, confusion_matrix
from sklearn.naive_bayes import GaussianNB  # Naive Bayes classifier
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
data = pd.read_csv("Intermediate result1.csv")

# Preprocessing the data
data = data.loc[:, ~data.columns.str.contains('^Unnamed')].dropna()

# Standardize features
scaler = StandardScaler()

# Function to handle target column and check class distribution
def train_and_evaluate(target_column, data, model):
    label_encoder = LabelEncoder()
    
    # Encode the target variable
    data[target_column] = label_encoder.fit_transform(data[target_column])
    
    # Check class distribution
    class_counts = data[target_column].value_counts()
    if any(class_counts < 2):
        print(f"Warning: {target_column} has classes with less than 2 samples. Adjusting dataset.")
        # Remove rows with rare classes
        valid_classes = class_counts[class_counts >= 2].index
        data = data[data[target_column].isin(valid_classes)]
    
    # Select features and target
    X = data.drop(columns=["Roll", target_column], errors="ignore")
    y = data[target_column]
    
    # Encode non-numeric columns
    non_numeric_cols = X.select_dtypes(include=["object"]).columns
    for col in non_numeric_cols:
        try:
            X[col] = label_encoder.fit_transform(X[col])
        except ValueError:
            X = X.drop(columns=[col])
    X_scaled = scaler.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.3, random_state=42, stratify=y
    )
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average="weighted")
    kappa = cohen_kappa_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average="weighted", zero_division=0)
    recall = recall_score(y_test, y_pred, average="weighted", zero_division=0)
    cm = confusion_matrix(y_test, y_pred)
    return accuracy, f1, kappa, precision, recall, cm
models = {
    "Naive Bayes": GaussianNB(),
    "MLP Classifier": MLPClassifier(hidden_layer_sizes=(50,), max_iter=500, random_state=42)
}
results = {}
targets = ["HSC GPA", "Proposed HSC  GPA-1", "Proposed HSC GPA-2"]
for target in targets:
    target_results = {}
    for model_name, model in models.items():
        target_results[model_name] = train_and_evaluate(target, data, model)
    results[target] = target_results
for target, model_results in results.items():
    print(f"Results for {target}:")
    for model_name, metrics in model_results.items():
        print(f"\n{model_name}:")
        print(f"  Accuracy: {metrics[0]:.2f}")
        print(f"  F1 Score: {metrics[1]:.2f}")
        print(f"  Cohen's Kappa: {metrics[2]:.2f}")
        print(f"  Precision: {metrics[3]:.2f}")
        print(f"  Recall: {metrics[4]:.2f}")
        print()
```
